```python
# 실제 train data를 변환하기 - 데이터 읽어들이기

import json
import os
import pandas as pd
import numpy as np

base_dir = 'C:/python-project/AI'

train_images_path = base_dir + '/train2/images/'
train_labels_json_path = base_dir + '/train2/labels_json/'
train_labels_path = base_dir + '/train2/labels/'

os.makedirs(train_images_path, exist_ok=True)
os.makedirs(train_labels_json_path, exist_ok=True)
os.makedirs(train_labels_path, exist_ok=True)

!unzip -qq C:/python-project/AI/train.zip -d C:/python-project/AI/train2/

train_images_list = os.listdir(train_images_path)
train_labels_json_list = os.listdir(train_labels_json_path)

train_images_list.sort()
train_labels_json_list.sort()
```

```python
# 실제 train data를 변환하기 - 라벨링 변환

yolo_class = {
    '철캔': 0,
    '알루미늄캔': 4,
    '종이': 8,
    '무색단일': 12,
    '유색단일': 16,
    'PE': 20,
    'PP': 24,
    'PS': 28,
    '스티로폼': 32,
    '비닐': 36,
    '갈색': 40,
    '녹색': 44,
    '투명': 48
}

labels_path0 = base_dir + '/data2/labels0/'
labels_path1 = base_dir + '/data2/labels1/'
labels_path2 = base_dir + '/data2/labels2/'
labels_path3 = base_dir + '/data2/labels3/'

images_path0 = base_dir + '/data2/images0/'
images_path1 = base_dir + '/data2/images1/'
images_path2 = base_dir + '/data2/images2/'
images_path3 = base_dir + '/data2/images3/'

os.makedirs(labels_path0, exist_ok=True)
os.makedirs(labels_path1, exist_ok=True)
os.makedirs(labels_path2, exist_ok=True)
os.makedirs(labels_path3, exist_ok=True)

os.makedirs(images_path0, exist_ok=True)
os.makedirs(images_path1, exist_ok=True)
os.makedirs(images_path2, exist_ok=True)
os.makedirs(images_path3, exist_ok=True)

import shutil
from PIL import Image

img_resize = 640

for label in train_labels_json_list:
    file_json = open(train_labels_json_path + label, encoding='UTF-8')
    f_json = json.load(file_json)
    file_json.close()
    img_width = f_json['IMAGE_INFO']['IMAGE_WIDTH']
    img_height = f_json['IMAGE_INFO']['IMAGE_HEIGHT']

    for anno in f_json['ANNOTATION_INFO']:
        obj_id = yolo_class[anno['DETAILS']]

        if anno['DAMAGE'] != '원형':
            obj_id += 2
        if anno['DIRTINESS'] != '오염없음':
            obj_id += 1
        
        image = label.replace('.json', '.jpg')
        
        if obj_id % 4 == 0:
            shutil.copy(train_labels_json_path + label, labels_path0 + label)
            image_ori = Image.open(train_images_path + image)
            image_tra = image_ori.resize((img_resize,img_resize))
            image_tra.save(images_path0 + image)
        elif obj_id % 4 == 1:
            shutil.copy(train_labels_json_path + label, labels_path1 + label)
            image_ori = Image.open(train_images_path + image)
            image_tra = image_ori.resize((img_resize,img_resize))
            image_tra.save(images_path1 + image)
        elif obj_id % 4 == 2:
            shutil.copy(train_labels_json_path + label, labels_path2 + label)
            image_ori = Image.open(train_images_path + image)
            image_tra = image_ori.resize((img_resize,img_resize))
            image_tra.save(images_path2 + image)
        else:
            shutil.copy(train_labels_json_path + label, labels_path3 + label)
            image_ori = Image.open(train_images_path + image)
            image_tra = image_ori.resize((img_resize,img_resize))
            image_tra.save(images_path3 + image)
```
